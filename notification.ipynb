```python
from winotify import Notification, audio
import time

def show_multiple_notifications():
    # Notification 1 - Unique App ID
    toast1 = Notification(
        app_id="My App 1",  # Different app_id for each notification
        title="Download Complete",
        msg="File 1 saved successfully!",
        duration="long"  # Keep open longer
    )
    toast1.show()
    time.sleep(1)  # Small delay between creations
    
    # Notification 2 - Different App ID
    toast2 = Notification(
        app_id="My App 2",  # Different app_id
        title="Download Complete", 
        msg="File 2 saved successfully!",
        duration="long"
    )
    toast2.show()
    time.sleep(1)
    
    # Notification 3 - Different App ID  
    toast3 = Notification(
        app_id="My App 3",  # Different app_id
        title="Download Complete",
        msg="File 3 saved successfully!",
        duration="long"
    )
    toast3.show()

show_multiple_notifications()
```

```python
from winotify import Notification
import threading
import time

def show_notification_thread(app_id, title, message, delay=0):
    time.sleep(delay)
    toast = Notification(
        app_id=app_id,
        title=title,
        msg=message,
        duration="long"
    )
    toast.show()

# Launch notifications with staggered timing
threading.Thread(target=show_notification_thread, 
                args=("App1", "Download 1", "File 1 complete", 0)).start()
threading.Thread(target=show_notification_thread, 
                args=("App2", "Download 2", "File 2 complete", 1)).start()
threading.Thread(target=show_notification_thread, 
                args=("App3", "Download 3", "File 3 complete", 2)).start()
```

```python
from winotify import Notification
import time

def show_stacked_notifications():
    notifications = [
        {"app_id": "DownloadManager1", "title": "Download Complete", "msg": "File 1 saved successfully!"},
        {"app_id": "DownloadManager2", "title": "Download Complete", "msg": "File 2 saved successfully!"},
        {"app_id": "DownloadManager3", "title": "Download Complete", "msg": "File 3 saved successfully!"}
    ]
    
    for i, notif_data in enumerate(notifications):
        toast = Notification(
            app_id=notif_data["app_id"],
            title=notif_data["title"],
            msg=notif_data["msg"],
            duration="short"  # Important: Keep notifications open
        )
        toast.show()
        
        # Wait for Windows to process the notification
        if i < len(notifications) - 1:  # Don't wait after the last one
            time.sleep(1)  # 2-second gap between notifications

show_stacked_notifications()
```

```python
from winotify import Notification
import threading
import time

def show_notification(app_id, title, message):
    toast = Notification(
        app_id=app_id,  # Different for each notification
        title=title,
        msg=message,
        duration="long"
    )
    toast.show()

# Launch both notifications at almost the same time
thread1 = threading.Thread(target=show_notification, 
                          args=("App1", "First Notification", "This is notification 1"))
thread2 = threading.Thread(target=show_notification, 
                          args=("App2", "Second Notification", "This is notification 2"))

thread1.start()
thread2.start()

# Wait for both to complete
thread1.join()
thread2.join()
```

```python
from winotify import Notification
import time

def show_two_notifications():
    # First notification
    toast1 = Notification(
        app_id="FirstApp",
        title="Notification 1",
        msg="First message here",
        duration="long"
    )
    toast1.show()
    
    # Very short delay
    time.sleep(0.5)  # Half second gap
    
    # Second notification  
    toast2 = Notification(
        app_id="SecondApp", 
        title="Notification 2",
        msg="Second message here",
        duration="long"
    )
    toast2.show()

show_two_notifications()
```

```python
from winotify import Notification
import threading

def dual_notification(title1, msg1, title2, msg2):
    def create_notification(app_id, title, message):
        Notification(
            app_id=app_id,
            title=title,
            msg=message,
            duration="long"
        ).show()
    
    # Launch both simultaneously
    t1 = threading.Thread(target=create_notification, args=("Notif1", title1, msg1))
    t2 = threading.Thread(target=create_notification, args=("Notif2", title2, msg2))
    
    t1.start()
    t2.start()
    t1.join()
    t2.join()

# Usage
dual_notification(
    "Download Complete", 
    "File 1 has been downloaded",
    "Process Finished", 
    "Background task completed"
)
```

```python
from winotify import Notification
import time

def show_stacked_notifications():
    # First notification - appears at bottom
    toast1 = Notification(
        app_id="BottomApp",
        title="First Notification", 
        msg="This appears first at the bottom",
        duration="long"
    )
    toast1.show()
    
    # Wait for first to fully appear
    time.sleep(3)  # 3-second delay
    
    # Second notification - appears on top of first
    toast2 = Notification(
        app_id="TopApp", 
        title="Second Notification",
        msg="This appears on top of the first one!",
        duration="long"
    )
    toast2.show()

show_stacked_notifications()
```

```python
from winotify import Notification
import time

def show_overlapping_notifications():
    # First notification
    Notification(
        app_id="BackgroundNotif",
        title="Background Notification",
        msg="I'll be covered by the next one",
        duration="long"
    ).show()
    
    # Shorter delay for quicker stacking
    time.sleep(2)
    
    # Second notification (will appear above first)
    Notification(
        app_id="ForegroundNotif", 
        title="Foreground Notification", 
        msg="I'm covering the previous notification!",
        duration="long"
    ).show()

show_overlapping_notifications()
```

```python
from winotify import Notification
import time

def show_corner_notifications():
    """
    Notifications will stack in bottom-right corner
    """
    positions = ["bottom-right", "top-right", "bottom-left"]  # Conceptual
    
    for i, pos in enumerate(positions):
        toast = Notification(
            app_id=f"PosTest{i}",
            title=f"Notification from {pos}",
            msg=f"This would be in {pos} if possible",
            duration="long"
        )
        toast.show()
        time.sleep(2)

show_corner_notifications()
```

```python
from winotify import Notification
import time

def controlled_timing_notifications():
    """
    Control timing and order instead of position
    """
    # First - appears lower in stack
    Notification(
        app_id="First",
        title="First in Stack", 
        msg="I'll be at the bottom",
        duration="long"
    ).show()
    
    time.sleep(2)
    
    # Second - appears above first
    Notification(
        app_id="Second",
        title="Second in Stack", 
        msg="I'll be on top of the first",
        duration="long" 
    ).show()
    
    time.sleep(2)
    
    # Third - appears at the top
    Notification(
        app_id="Third",
        title="Third in Stack",
        msg="I'm at the very top!",
        duration="long"
    ).show()

controlled_timing_notifications()
```

```python
# Use tkinter for custom positioned windows instead
import tkinter as tk
from tkinter import messagebox

def create_custom_popup():
    root = tk.Tk()
    root.withdraw()  # Hide main window
    
    # Create popup at specific position
    root.geometry("+100+100")  # x=100, y=100
    messagebox.showinfo("Custom Position", "This appears at 100,100", parent=root)
    root.destroy()

create_custom_popup()
```
